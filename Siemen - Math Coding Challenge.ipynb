# 2. Rishi has a total of  590 as currency notes in the denominations of 50,  20 and  10. The ratio of the number of  50 notes and  20 notes is 3:5. If she has a total of 25 notes, how many notes of each denomination she has?

# Solution:

```python
Total_rupees = 590
r50 = 3 
r20 = 5 
x = symbols ('x') 

eqs = 50 * r50 * x + 20 * r20 * x + (25 - (8 * x)) * 10 - 590

sol = solve ((eqs), (x)) 

note50=r50*sol[0]
note20=r20*sol[0]
note10=(25 - (8 * sol[0]))
```

```python
print (f'1. {note50} notes of 50\n2. {note20} notes of 20\n3. {note10} notes of 10')
```

# 3. The organizers of an essay competition decide that a winner in the competition gets a prize of 100 and a participant who does not win gets a prize of 25. The total prize money distributed is  3,000. Find the number of winners, if the total number of participants is 63.

# Solution:

```python
win_prize_money = 100
participant_money = 25
total_money = 3000
total_participant = 63

x=symbols('x')
 
eq=win_prize_money*x+(total_participant-x)*participant_money-total_money
sol=solve(eq,x)
```

```python
print(f'Total {sol[0]} participants won the competition')
```

# 4. The sum of three consecutive multiples of 11 is 363. Find these multiples.

```python
x=symbols('x')

first=11*x
second=11*(x+1)
third=11*(x+2)

eq=first+second+third-363

sol=solve(eq,x)
```

```python
print(f'First multiple : {11*sol[0]} \nSecond multiple : {11*sol[0]+11} \nThird multiple : {11*sol[0]+22}')
```

# 5. The digits of a two-digit number differ by 3. If the digits are interchanged, and the resulting number is added to the original number, we get 143. What can be the original number?

# Solution:

```python
a,b=symbols('a b')

eq1=a-b-3

eq2=10*a+b+10*b+a-143

sol=solve((eq1,eq2),(a,b))
```

```python
print(f'The required number is {10*sol[a]+sol[b]} or {sol[b]*10+sol[a]}')
```

# 7. Find the area of the region included between the parabola y2 = x and the line x + y = 2 and the X-axis.

# Solution:

```python
def integrand(x):
    return 2-x-x**2

ans, err = quad(integrand, -2, 1)
```

```python
print(f"Area : {ans}")
```

# 8.  A manufacturer produces three models of toys in the form of bikes say X,Y and Z. Model X takes as 10 man-hour to make per unit, Model Y takes 5 man-hour per unit and model Z takes 4 man-hour per unit. There are a total 212 man-hour available per week. Handling and marketing costs are <20, <30 and <40 per unit for models X,Y and Z respectively. The total funds available for these purposes are <920 per week. Profits per unit for models X,Y and Z are <40, <10and <70 respectively, but at the end of the week, company get a profit of <810. Solve the system of equations by matrix method.

# Solution:

```python
# define matrix A using Numpy arrays
A = np.array([[10, 5, 4],
             [20, 30, 40],
             [40, 10, 70]]) 

#define matrix B
B = np.array([212, 920, 810]) 
```

```python
# linalg.solve is the function of NumPy to solve a system of linear scalar equations
print ("Solution:\n",np.linalg.solve(A,B ))
```
